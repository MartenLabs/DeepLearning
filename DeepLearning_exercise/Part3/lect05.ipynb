### bar plot

bar plot은 ax.bar와 같이 사용하며 다음과 같은 인자를 받습니다.

- x: x축에 사용된 값
- height: y축에 사용된 값
- width : 각 bar의 width
- bottom: y축의 base값
- align : 정렬(center || edge)

```python
# bar plot
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6]) # 기본 width = 0.8
plt.show()
```

```python
# width 조정 
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], width=0.5)
plt.show()
```

```python
# bottom 조절
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], bottom=2) # y축의 시작을 2로 설정
plt.show()
```

```python
# align 조절
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], align='edge') # bar의 끝을 기준으로 정렬
plt.show()
```

### histogram plot

histogram은 data의 분포를 나타낼 때 사용하며, 주로 다음과 같은 인자를 사용합니다.

- x    : 입력 data
- bins : histogram을 나타낼 때 x축의 구간(정수 입력시 구간의 갯수, list와 같은 sequence 입력시는 sequence 값이 구간으로 설정됨)
- range: x축의 범위(x.min, x.max)로 표현
- density: 확률분포로 나타낼 것인지 여부

```python
# histogram
np.random.seed(1)

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500))
plt.show()
```

```python
# bins 조절
np.random.seed(1)

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500), bins=50) # 50개의 막대바 생성 
plt.show()
```

```python
# bins 조절2
np.random.seed(1)

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500), bins=[-3, -2, -1, 1, 2, 3]) # -3 ~ 3 까지 6개의 막대바 생성
plt.show()
```

```python
# range 조절
np.random.seed(1)

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500), bins = 50, range=(-3, 3)) # -3 ~ 3 까지 50개의 막대바 생성
plt.show()
```

```python
# density 조절
np.random.seed(1)

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500), bins = 50, range = (-2, 2), density=True) # 확률분포로 표현
plt.show()
```

```python
# subplot 간격 조정
np.random.seed(1)

fig, axes = plt.subplots(2, 2, figsize=(10, 5), sharey=True, tight_layout=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50)

plt.subplots_adjust(wspace = 0, hspace = 0)
plt.show()
```

### label 지정하기

### 제목, 축이름, 눈금 설정하기

```python
fig = plt.figure(figsize=(7, 7))

ax = fig.add_subplot()

for species in iris_df['Species'].unique():
    iris_sub = iris_df[iris_df['Species'] == species]
    ax.scatter(x = iris_sub['PetalLengthCm'],
               y = iris_sub['PetalWidthCm'],
               label = species)
    
ax.set_title('Iris Data')
xticks = np.arange(0, 7.5, 0.5) # x축 눈금(0 ~ 7.5 까지 0.5 간격)
ax.set_xticks(xticks) # x축 눈금
ax.set_xlabel('PetalLengthCm') # x축 라벨
ax.set_ylabel('PetalWidthCm') # y축 라벨

plt.legend()
plt.show()
```

Oxford IIIT Pet dataset을 다운받아서 화면에 출력해봅시다.

```python
import os
import random
import xml.etree.ElementTree as et
from matplotlib.patches import Rectangle
```

```python
import gdown
url = 'https://drive.google.com/uc?id=1dIR9ANjUsV9dWa0pS9J0c2KUGMfpIRG0'
fname = 'data/oxford_pet.zip'
gdown.download(url, fname, quiet=False)
```

```python
## 압축 풀기
!unzip -q data/oxford_pet.zip -d oxford_pet
```

```python
cur_dir = os.getcwd()
data_dir = os.path.join(cur_dir, 'oxford_pet')
image_dir = os.path.join(data_dir, 'images')
bbox_dir = os.path.join(data_dir, 'annotations', 'xmls')
seg_dir = os.path.join(data_dir, 'annotations', 'trimaps')
```

```python
image_files = [fname for fname in os.listdir(image_dir) if os.path.splitext(fname)[-1] == '.jpg']
bbox_files = [fname for fname in os.listdir(bbox_dir) if os.path.splitext(fname)[-1] == '.xml']
seg_files = [fname for fname in os.listdir(seg_dir) if os.path.splitext(fname)[-1] == '.png']
```

```python
rnd_idx = random.randint(1, len(bbox_files))

bbox_name = bbox_files[rnd_idx]
bbox_path = os.path.join(bbox_dir, bbox_name)
tree = et.parse(bbox_path)

width = float(tree.find('./size/width').text)
height = float(tree.find('./size/height').text)
xmin = float(tree.find('./object/bndbox/xmin').text)
xmax = float(tree.find('./object/bndbox/xmax').text)
ymin = float(tree.find('./object/bndbox/ymin').text)
ymax = float(tree.find('./object/bndbox/ymax').text)

rect_x = xmin
rect_y = ymin
rect_w = xmax-xmin
rect_h = ymax-ymin

fname = os.path.splitext(bbox_name)[0]+'.jpg'
fpath = os.path.join(image_dir, fname)
image = Image.open(fpath)
# image = np.array(image)

bbox_name = os.path.splitext(fname)[0]+'.xml'

fig, ax = plt.subplots(1, 1)
rect = Rectangle((rect_x, rect_y), rect_w, rect_h, fill=False, color='red')
ax.add_patch(rect)
ax.imshow(image)
plt.show()
```

```python
sname = os.path.splitext(bbox_name)[0]+'.png'
spath = os.path.join(seg_dir, sname)
seg = Image.open(spath)

fig, ax = plt.subplots(1, 2, figsize=(15,8))
ax[0].imshow(image)
ax[1].imshow(seg)
plt.show()
```
