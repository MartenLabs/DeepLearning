### Odds 그래프와 Logit 그래프를 pyplot을 이용하여 그리시오

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

p_np = np.linspace(0.01, 0.99, 100)
p_tf = tf.linspace(0.01, 0.99, 100)

np_odds = p_np / (1 - p_np)
tf_odds = p_tf / (1 - p_tf)

np_Logit = np.log(np_odds)
tf_Logit = tf.math.log(tf_odds)


fig, axis = plt.subplots(2, 1, figsize = (15, 10), sharex = True)

axis[0].plot(p_np, np_odds)
axis[1].plot(p_np, tf_Logit.numpy())

xticks = np.arange(0, 1.1, 0.1)

axis[0].tick_params(labelsize = 15)
axis[0].set_xticks(xticks)
axis[0].set_ylabel('Odds', fontsize = 20, color = 'darkblue')

axis[1].tick_params(labelsize = 15)
axis[1].set_ylabel('Logit', fontsize = 20, color = 'darkblue')
axis[1].set_xlabel('Probability', fontsize = 20, color = 'darkblue')
```

### sigmoid 그래프를 pyplot을 사용하여 그리시오

```python
import matplotlib.pyplot as plt
import tensorflow as tf 
from tensorflow.keras.layers import Activation

X = tf.linspace(-10, 10, 100)
sigmoid = Activation('sigmoid')(X)

fig, ax = plt.subplots(figsize = (10, 5))
ax.plot(X.numpy(), sigmoid.numpy())
```

### 100개의 normal distribution을 따르는 랜덤 number를 sigmoid함수를 통과 한 뒤의 그래프를 pyplot을 사용하여 그리시오

```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Dense 

X = tf.random.normal(shape = (100, 1))
dense = Dense(units = 1, activation = 'sigmoid')
Y = dense(X)

print(Y.shape)

fig, ax = plt.subplots(figsize = (15, 10))
ax.scatter(X.numpy().flatten(), Y.numpy().flatten())
```

### Binary Classifier를 수행하는 3-Dense Layer를 구현하시오

dense1 = (10, relu)

dense2 = (5, relu)

dense3 = (3, sigmoid)

input x

```python
import tensorflow as tf 
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential

N = 100
n_feature = 10

X = tf.random.normal(shape = (N, 10))
model = Sequential()

model.add(Dense(units = 10, activation = 'relu'))
model.add(Dense(units = 5, activation = 'relu'))
model.add(Dense(units = 3, activation = 'sigmoid'))

Y = model(X)
print(Y.shape)
```

### 8개의 class를 분류 할 수 있는 1 dense layer를 구현하시오
input = (8, 5) uniform distribution

```python
import tensorflow as tf 
from tensorflow.keras.layers import Dense 

logit = tf.random.uniform(shape = (8, 5), minval = -10, maxval = 10)

dense = Dense(units = 8, activation = 'softmax')
Y = dense(logit)
print(tf.reduce_sum(Y,axis = 1))
```

### Multi-calss Classifiers 3-dense layer를 class로 구현하시오

dense = 8, 5, 3 

input = (8, 5) uniform distribution을 따르며 min = -10, max = 10


```python
import tensorflow as tf 
from tensorflow.keras.layers import Dense 
from tensorflow.keras.models import Model

class TestClass(Model):
    def __init__(self):
        super(TestClass, self).__init__()
        
        self.dense1 = Dense(units = 8, activation = 'sigmoid')
        self.dense2 = Dense(units = 5, activation = 'sigmoid')
        self.dense3 = Dense(units = 3, activation = 'softmax')
        
    def __call__(self, x):
        x = self.dense1(x)
        x = self.dense2(x)
        x = self.dense3(x)
        
        return x
    
N, n_feature = 8, 5
x = tf.random.uniform(shape = (N, n_feature), minval = -10, maxval = 10)
model = TestClass()
Y = model(x)
print(tf.reduce_sum(Y, axis = 1))
print(Y.shape)
```
